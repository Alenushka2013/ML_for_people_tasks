1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import pandas as pd
import numpy as np
import sympy as sp

import matplotlib.pyplot as plt
```

```python
def calc_f_x(func, x):
  result = func(x)
  # Якщо результат — константа, повторити його для кожного елемента x
  if np.isscalar(result):
    return np.full_like(x, result, dtype=float)
  return result

def deriv_and_plot(f, a, b):
  """
  Обчислення похідної для заданої функції f
  Побудова графіків функції та похідної на проміжку [a, b]
  """
  # Визначаємо символічну змінну
  x = sp.symbols('x')

  # Знаходження похідної
  deriv_f = sp.diff(f, x)

  # Перетворення символьного виразу на числову функцію
  f_func = sp.lambdify(x, f, 'numpy')
  deriv_f_func = sp.lambdify(x, deriv_f, 'numpy')

  x = np.linspace(a, b, 100)

  # Створення графіку
  plt.figure(figsize=(10, 5))
  plt.plot(x, calc_f_x(f_func, x), label=f'f(x) = {f}', color='blue')
  plt.plot(x, calc_f_x(deriv_f_func, x), label=f"f'(x) = {deriv_f}", color='green')

  plt.title(f"Графік функції f(x) = {f} та її похідної f'(x) = {deriv_f} на інтервалі [{a}, {b}]")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.legend()
  plt.grid(True)

  plt.show()
```

```python
# Визначаємо символічну змінну
x = sp.symbols('x')

f = 10 * x + 5
a, b = 0, 10

deriv_and_plot(f, a, b)
```

Для функції f(x)=10x−5 похідна f'(x)  = 10.

З кожною зміною значення аргументу на одиницю значення функції збільшується на 10.

Тобто з кожною годиною витраченою на навчання оцінка виростає на 10 одиниць.
